# Gaussian Blob Dataset – Examples

Visualize a few samples from the `GaussianBlobDataset`.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt
from src.datasets.gaussian_blob import GaussianBlobDataset
```

```python
ds = GaussianBlobDataset(n_train=4, n_test=0)
data = ds.create()

images = data["train_input"].reshape(-1, 10, 10)
labels = data["train_label"]

# un-normalize labels back to original scale
x_center = labels[:, 0] * 9
y_center = labels[:, 1] * 9
width    = labels[:, 2] * 2.0 + 0.5
amplitude = labels[:, 3] * 0.8 + 0.2
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 3.5))

for i, ax in enumerate(axes):
    im = ax.imshow(images[i], cmap="viridis", origin="lower", extent=[0, 9, 0, 9])
    ax.plot(x_center[i], y_center[i], "r+", markersize=12, markeredgewidth=2)
    ax.set_title(
        f"x={x_center[i]:.1f}  y={y_center[i]:.1f}\n"
        f"w={width[i]:.2f}  amp={amplitude[i]:.2f}"
    )
    fig.colorbar(im, ax=ax, shrink=0.8)

fig.suptitle("Gaussian Blob Samples", fontweight="bold", y=1.02)
fig.tight_layout()
plt.show()
```
